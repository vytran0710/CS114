# Import

```
import numpy as np
import os
from sklearn.svm import SVC
from sklearn.svm import LinearSVC
from sklearn import metrics
from sklearn.metrics import accuracy_score
```

# Load data - Training set, testing set and validation set

```
path_dir = '/content/drive/My Drive/Hoc/CS114/Dataset/Food-11/features/VGG16_fc2/'
os.chdir(path_dir)
evaluation_dir = path_dir + '/evaluation/'
training_dir = path_dir + '/training/'
validation_dir = path_dir + '/validation/'
```

```
X_train = []
Y_train = []
```

```
lables_training = os.listdir(training_dir)
for i in lables_training:
  lables_training_dir = training_dir + i + '/'
  print(i)
  for file in os.listdir(lables_training_dir):
    f = lables_training_dir + file 
    print(f)
    X_train.append(np.load(f))
    Y_train.append(i)
  print('Complete loading', i, 'class (Training) ...\n')
```

```
X_test = []
Y_test = []
```

```
lables_evaluation = os.listdir(evaluation_dir)
for i in lables_evaluation:
  lables_evaluation_dir = evaluation_dir + i + '/'
  print(i)
  for file in os.listdir(lables_evaluation_dir):
    f = lables_evaluation_dir + file 
    print(f)
    X_test.append(np.load(f))
    Y_test.append(i)
  print('Complete loading', i, 'class (Evaluation) ...\n')
```

# OvR

```
OvR = SVC(kernel='linear', decision_function_shape = "ovr")
OvR.fit(X_train, Y_train)
Y_pred_OvR = OvR.predict(X_test)
print('Accuracy of OVR:', metrics.accuracy_score(Y_pred_OvR, Y_test))
print('Shape of OVR:', OvR.support_vectors_.shape)
print('Number of support vectors for each class:', OvR.n_support_)
```

# OvO

```
OvO = SVC(kernel='linear', decision_function_shape = 'ovo')
OvO.fit(X_train, Y_train)
Y_pred_OvO = OvO.predict(X_test)
print('Accuracy of OvO:', metrics.accuracy_score(Y_pred_OvO, Y_test))
print('Shape of OvO:', OvO.support_vectors_.shape)
print('Number of support vectors for each class:', OvO.n_support_)
```

# LinearSVC

```
LSVC = LinearSVC(dual = False)
LSVC.fit(X_train, Y_train)
Y_pred_LinearSVC = LSVC.predict(X_test)
print('Accuracy of LinearSVC:', metrics.accuracy_score(Y_pred_LinearSVC, Y_test))
print('Shape of LinearSVC:', LSVC.coef_.shape)
```

# OvR, OvO, LinearSVC results

```
print("-----------------------------------------------------------------------------------------------------------------")
print("|\t\t|\t SVC (OvO) \t\t|\t SVC (OvR) \t\t|\t LinearSVC \t\t|")
print("-----------------------------------------------------------------------------------------------------------------")
print("|\tAcc\t|\t",metrics.accuracy_score(Y_pred_OvO, Y_test),"\t|\t",metrics.accuracy_score(Y_pred_OvR, Y_test),"\t|\t",metrics.accuracy_score(Y_pred_LinearSVC, Y_test),"\t|")
print("-----------------------------------------------------------------------------------------------------------------")
print("|\tSize\t|\t",OvO.support_vectors_.shape,"\t\t|\t",OvR.support_vectors_.shape,"\t\t|\t",LSVC.coef_.shape,"\t\t|")
print("-----------------------------------------------------------------------------------------------------------------")
```

# KNN

```
from sklearn.neighbors import KNeighborsClassifier
for i in range(1, 11):
  KNN = KNeighborsClassifier(n_neighbors = i)
  KNN.fit(X_train, Y_train)
  Y_pred_KNN = KNN.predict(X_test)
  print('Accuracy of k =', i, ":\t", metrics.accuracy_score(Y_pred_KNN, Y_test))
```

# Bayes - Bernoulli

```
from sklearn.naive_bayes import BernoulliNB
BNB = BernoulliNB()
BNB.fit(X_train, Y_train)
Y_pred_BNB = BNB.predict(X_test)
print('Accuracy of Bayes - Bernoulli:', metrics.accuracy_score(Y_pred_BNB, Y_test))
print('Size:', BNB.feature_count_.shape)
```
